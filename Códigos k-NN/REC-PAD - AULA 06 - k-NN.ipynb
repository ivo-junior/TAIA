```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('iris.csv', header=None)
iris

# Cabeçalho:
# id | SepalLengthCm | SepalWidthCm | PetalLengthCm | PetalWidthCm | Species
```

```python
from sklearn.preprocessing import OrdinalEncoder
import copy

# Define atributos como 4, coluna onde as classes são armazenadas na Iris
classes = [4]
print(iris[classes])
enc = OrdinalEncoder()

# "Encaixa" nosso codificador nos valores presentes em iris[atributos]
enc.fit(iris[classes])
# Transforma os valores de iris[atributos] nos valores codificados, que são inteiros representando cada classe
iris[classes] = enc.transform(iris[classes])
print(iris[classes])

# Separa os atributos X, das classificações Y
X = iris.iloc[:,:4]
print(X)
y = iris.iloc[:, 4]
print(y)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Usa PCA para reduzir a quantidade de atributos de 4 para 2
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X.values)

print('Projecting %d-dimensional data to 2D' % X.values.shape[1])

plt.figure(figsize=(12,10))
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y.values, 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 3))
plt.colorbar()
plt.title('PCA projection');
```

# Utilização do k-NN

### Sem PCA

```python
iris = pd.read_csv('iris.csv', header=None)

# Divisão em atributos e classes
X = iris.iloc[:,:4]
y = iris[4]

# Vamos remover a linha 149 do nosso conjunto de treinamento
# E usá-la como teste do nosso modelo k-NN
X_teste = X.iloc[149]
X = X.drop(149)
y_teste = y.iloc[149]
y = y.drop(149)
y_teste
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Vamos utilizar os três vizinhos mais próximos para prever nossa classe.
knn = KNeighborsClassifier(n_neighbors=3)

# Executar o treinamento
knn.fit(X, y)
# Vamos testar na linha 149
knn.predict(np.array(X_teste).reshape(1,-1))
```

```python
dist, vizinhos = knn.kneighbors(np.array(X_teste).reshape(1,-1))
classes = []
for v in vizinhos:
    classes.append(iris.iloc[:-1, 4][v])
classes
```

# Com conjunto de testes

```python
import numpy as np
import pandas as pd
iris = pd.read_csv('iris.csv', header=None)

# Divisão em atributos e classes
X = iris.iloc[:,:4]
y = iris[4]

# Vamos remover 20% do nosso conjunto de treinamento
# E usar como teste do nosso modelo k-NN
list_ids = list(range(len(y)))
ids_test = list(np.random.choice(list_ids, int(len(list_ids)*0.2), replace=False))

print(len(y), len(ids_test))
X_teste = X.iloc[ids_test]
X = X.drop(index=ids_test)
y_teste = y.iloc[ids_test]
y = y.drop(index=ids_test)
print(len(y), len(y_teste))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Vamos utilizar os 2 vizinhos mais próximos para prever nossa classe.
print(f"Testando os 3-NN")
knn = KNeighborsClassifier(n_neighbors=3)
# Executar o treinamento
knn.fit(X, y)
# Vamos testar na linha na nossa base de testes
preds = knn.predict(np.array(X_teste))
for p, gt in zip(preds,y_teste):
    print(f'{p}\t\t{gt}:    \t{p==gt}')
acertos = sum(preds == y_teste)
print('Resultado: ',acertos/len(y_teste)*100,'%')
print('==================')
```
