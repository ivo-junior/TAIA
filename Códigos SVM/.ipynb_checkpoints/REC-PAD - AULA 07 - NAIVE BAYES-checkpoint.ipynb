```python
import pandas as pd
import numpy as np
```

```python
wine = pd.read_csv('wine.csv')
wine
```

# Utilizando NB

```python
X = wine.iloc[:,1:]
y = wine.iloc[:, 0]

X_teste = X.iloc[177]
X = X.drop(177)
y_teste = y.iloc[177]
y = y.drop(177)

y_teste
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X, y)
gnb.predict(np.array(X_teste).reshape(1,-1))
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(X, y)
mnb.predict(np.array(X_teste).reshape(1,-1))
```
